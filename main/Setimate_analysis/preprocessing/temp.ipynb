```python
import numpy as np
import pandas as pd
import os
import re
```

```python
path = "../../artifacts/chat.csv"
data = pd.read_csv(path)
```

```python
data.dropna(inplace=True)
```

```python
# Extract text
text = ""
for i in data['messages']:
    text+=" "+i
text
```

```python
# Lowercassing
# text
text = text.lower()
text
```

```python
# Remove links
patt = r"http[s]?://\S+"
text = re.sub(patt,"",text)
```

```python
# remove emoji
# import emoji
# text = emoji.demojize(text)
```

```python
# Remove Numbers
text = re.sub(r"\d+", "", text)
```

```python
sentence_chat = text
```

```python
# Remove Symbols
text = re.sub(r'[^\w\s]|\n\n|\n|\\', ' ', text)
```

```python
# Tokenization
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from textblob import TextBlob

blob = TextBlob(text)
# print(blob.words)
```

```python
clean = blob.words
```

```python
# remove white space
clean = [w.strip() for w in clean]
```

```python
# Remove Duplicates

# text = set()
# for w in clean:
#     text.add(w)

# In one line code
text = {w for w in clean}
```

```python
# Remove stop word
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

text_str = " ".join(text)
tokens = word_tokenize(text_str)
```

```python
# Stemming
from nltk.stem import PorterStemmer

ps = PorterStemmer()
stemmed = [ps.stem(w) for w in tokens]
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(w, pos="v") for w in stemmed]
```

```python
len(lemmas)
```

```python
text = {w for w in lemmas}
text = list(text)
len(text)
```

```python
text
```

```python
# Spelling Correction
from textblob import TextBlob

text_str2 = " ".join(text)
blob = TextBlob(text_str2)
corrected = str(blob.correct())
```

```python
text = corrected.split()
text
```

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(sentence_chat)

new_sentence = []
for sentence in sentences:
    text = re.sub(r'[^\w\s]|\n\n|\n|\\', ' ', sentence)
    cleaned = re.sub(r'\s+', ' ', text).strip()
    if len(cleaned) !=0:
        new_sentence.append(cleaned)
new_sentence
```

```python
text_7 = []
for sentence in new_sentence:
    words = sentence.split()
    
    n=len(words)
    if n > 7 :
        while True:
            temp = words[:7]
            temp = " ".join(temp)
            text_7.append(temp)
            words = words[7:]
            if n>=7:
                n-=7
            elif n>0 and n<7:
                n = -1
                continue
            else:
                break
    else:
        words = " ".join(words)
        text_7.append(sentence)
       
```

```python
import nltk
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download VADER if not already
nltk.download('vader_lexicon')

# Join tokens back into a sentence
chat_text = " ".join(text)

# Initialize VADER
sia = SentimentIntensityAnalyzer()

# Function to get sentiment
def get_sentiment(text):
    score = sia.polarity_scores(text)['compound']
    if score > 0.05:
        return "positive"
    elif score < -0.05:
        return "negative"
    else:
        return "neutral"

# Apply sentiment labeling
labeled_data = [(msg, get_sentiment(msg)) for msg in text_7]

# Convert to DataFrame
df = pd.DataFrame(labeled_data, columns=["text", "sentiment"])
df.dropna(inplace=True)

# Save to CSV
df.to_csv("chat_labeled.csv", index=False, encoding="utf-8")

print(df)
print("\nâœ… Saved to chat_labeled.csv")
```
